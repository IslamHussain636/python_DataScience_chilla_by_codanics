# Data Normalization
- uniform data 
- same impact
- bell curved uniform


````python
#kashti
kashti.head(5)
````

````python
ks11= kashti[['age in days','fare']]
ks11.head(5)
````

````python
# wide range data unable to plot in proper 
#methods of normalization 
````

````python
# wide range data unable to plot in proper 
#methods of normalization 
#simple feature scaling 
ks11['fare']=ks11['fare']/ks11['fare'].max()
ks11['age in days']=ks11['age in days']/ks11['age in days'].max()
ks11.head(5)
````

````python
# Min - Max method 
ks11['fare']=(ks11['fare']-ks11['fare'].min())/ks11['fare'].max()-ks11['fare'].min()
ks11['age in days']=(ks11['age in days']-ks11['age in days'].min())/ks11['age in days'].max()-ks11['age in days'].min()
ks11.head(5)
````

````python
# Z score range -3 to 3
ks11['fare']=(ks11['fare']-ks11['fare'].mean())/ks11['fare'].std()
ks11['age in days']=(ks11['age in days']-ks11['age in days'].mean())/ks11['age in days'].std()
ks11.head()
````

````python
# log transformation 
ks11['fare']=np.log(ks11['fare'])
ks11['age in days']=np.log(ks11['age in days'])
ks11.head(5)
````
